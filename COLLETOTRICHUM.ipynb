## BAR PLOT

### 1PR

```python
d_counts = df_1.groupby("SP")["Z"].apply(lambda x: x.eq("D").sum()).reset_index()
d_counts.plot(x="SP", kind="bar", width=0.8)
plt.legend(bbox_to_anchor=(1.2,1))
plt.show()
```

```python
d_counts = df_1.groupby("SP")["Z"].apply(lambda x: x.eq("D").sum()).reset_index()
s_counts = df_1.groupby("SP")["Z"].apply(lambda x: x.eq("S").sum()).reset_index()
ratio = d_counts["Z"]/(s_counts["Z"]+d_counts["Z"])
ratio.plot(x="SP", kind="bar", width=0.8)
```

```python
d_counts = df_2.groupby("SP")["AC"].apply(lambda x: x.eq("D").sum()).reset_index()
d_counts.plot(x="SP", kind="bar", width=0.8)
plt.legend(bbox_to_anchor=(1.2,1))
plt.show()
```
